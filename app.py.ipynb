```python
import os
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output

app = dash.Dash(__name__)
server = app.server
colors = {
    'background': '#111111',
    'text': '#7FDBFF',
    "subtext": "#FF9900"
}
app.layout = html.Div(style={"backgroundColor": colors["background"]}, children=[
    html.H1('Welcome to our app', style= {"textAlign" : "center", "color" : colors["text"]}),
    html.H2("Optimize scheduling", style= {"texteAlign" : "center", "color" : colors["subtext"]}),
    html.Label('Multi-Select Dropdown'),
    dcc.Dropdown(id='dropdown',
                           options= [{'label': i, 'value': i} for i in ['Machine 1', 'Machine 2', 'Machine 3']],
                           multi=True,
                           style={'backgroundColor': '#FFFFFF'},
                           ),
       
    html.Label('Checkboxes'),
    dcc.Checklist(
        options=[{'label': i, 'value': i} for i in ['Product 1', 'Product 2', 'Product 3', "Product 4"]],
        style={'backgroundColor': '#1E1E1E', "color" : "#FFFFFF"},
    ),
         
    html.Div(id='display-value', style = {"backgroundColor" : "#111111"}),
    dcc.Input(id='my-id', value='Enter your email adress', type='text'),
    html.Div(id='my-div', style = {"backgroundColor" : "#111111", "color" : "#FFFFFF"}),
    html.H2("Product 1", style= {"textAlign" : "left", "color" : colors["text"]}),
    dcc.Input(id='my-id2', value='Enter a number', type='text'),
    html.Div(id='my-div2', style = {"backgroundColor" : "#111111", "color" : "#FFFFFF"}),
    html.H2("Product 2", style= {"textAlign" : "left", "color" : colors["text"]}),
    dcc.Input(id='my-id3', value='Enter a number', type='text'),
    html.Div(id='my-div3', style = {"backgroundColor" : "#111111", "color" : "#FFFFFF"}),
    html.H2("Product 3", style= {"textAlign" : "justify", "color" : colors["text"]}),
    dcc.Input(id='my-id4', value='Enter a number', type='text'),
    html.Div(id='my-div4', style = {"backgroundColor" : "#111111", "color" : "#FFFFFF"}),     
    html.H2("Product 4", style= {"textAlign" : "justify", "color" : colors["text"]}),
    dcc.Input(id='my-id5', value='Enter a number', type='text'),
    html.Div(id='my-div5', style = {"backgroundColor" : "#111111", "color" : "#FFFFFF"}),
    html.Div(dcc.Input(id='input-on-submit', type='text')),
    html.Button('Submit', id='submit-val', n_clicks=0),
    html.Div(id='container-button-basic', children='Enter a value and press submit'),
    html.Button('RUN', id='btn-nclicks-1', n_clicks=0, style = {"backgroundColor": '#7FDBFF'})
    

])


    
app.css.append_css({"external_url": "https://codepen.io/chriddyp/pen/bWLwgP.css"})


#@app.callback(dash.dependencies.Output('display-value', 'children'), [dash.dependencies.Input('dropdown', 'value')])
#def display_value(value):
 #   return 'You have selected "{}"'.format(value)
@app.callback(
    Output(component_id='my-div', component_property='children'),
    [Input(component_id='my-id', component_property='value')]
)
def update_output_div(input_value):
    return 'You\'ve entered "{}"'.format(input_value)

@app.callback(
    Output(component_id='my-div2', component_property='children'),
    [Input(component_id='my-id2', component_property='value')]
)
def update_output_div(input_value):
    return 'You\'ve entered "{}"'.format(input_value)

@app.callback(
    Output(component_id='my-div3', component_property='children'),
    [Input(component_id='my-id3', component_property='value')]
)
def update_output_div(input_value):
    return 'You\'ve entered "{}"'.format(input_value)

@app.callback(
    Output(component_id='my-div4', component_property='children'),
    [Input(component_id='my-id4', component_property='value')]
)
def update_output_div(input_value):
    return 'You\'ve entered "{}"'.format(input_value)

@app.callback(
    Output(component_id='my-div5', component_property='children'),
    [Input(component_id='my-id5', component_property='value')]
)
def update_output_div(input_value):
    return 'You\'ve entered "{}"'.format(input_value)

@app.callback(
    dash.dependencies.Output('container-button-basic', 'children'),
    [dash.dependencies.Input('submit-val', 'n_clicks')],
    [dash.dependencies.State('input-on-submit', 'value')])
def update_output(n_clicks, value):
    return 'The input value was "{}" and the button has been clicked {} times'.format(
        value,
        n_clicks
    )

if __name__ == '__main__':
    app.run_server(debug=True)
```
